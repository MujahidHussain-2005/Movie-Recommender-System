```python
# columns we will keep
# genres
# id
# keywords
# overview   
# title 
# cast
# crew
```

```python
movies=movies[['genres','id','keywords','overview','title','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
# bring above format to below form
# Adventure,Fantasy Action
```

```python
import json
def convert(obj):
    list=[]
    for i in json.loads(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['cast'][0]
```

```python
def convert1(obj):
    list=[]
    check=1
    for i in json.loads(obj):
        if check<=3:
            list.append(i['name'])
            check+=1
        else:
            break
    return list
```

```python
movies['cast']=movies['cast'].apply(convert1)
```

```python
movies.head(2)
```

```python
# crew 
```

```python
movies['crew'][0]
```

```python
def convert2(obj):
    list=[]
    for i in json.loads(obj):
        if i['job']=='Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew']=movies['crew'].apply(convert2)
```

```python
movies.head(4)
```

```python
# overview
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
# space removal
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','')for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','')for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','')for i in x])
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
movies.head(3)
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
movies.head(3)
```

```python
df=movies[['id','title','tags']]
```

```python
df['tags']=df['tags'].apply(lambda x: ' '.join(x))
```

```python
df['tags']=df['tags'].apply(lambda x: x.lower() )
```

```python
df.head(2)
```

### Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(value):
    list1=[]
    for i in value.split():
        list1.append(ps.stem(i))
    return " ".join(list1)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df['tags'][0]
```

### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(df['tags']).toarray()
```

```python
vectors[0]
```

### Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
similarity.shape
type(similarity)
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
sorted(similarity[1],reverse=True)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
list(enumerate(sorted(similarity[1],reverse=True)))
```

```python
def recommend(movie):
    index=df[df['title'] == movie].index[0]
    distances=similarity[index]
    movies=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies:
        print(df.iloc[i[0]].title)
    
```

```python
import pickle
pickle.dump(df.to_dict(),open('df.pkl','wb'))
```

```python
pickle.dump(similarity,open('similarity.pkl','wb'))
```
